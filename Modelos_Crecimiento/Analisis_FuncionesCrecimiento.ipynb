# Modelos de crecimiento

En este notebook se simulara observaciones de los principales modelos de crecimiento, esto para familiarizarse y analizar sus principales caracteristicas

```python
# Definiendo librerias a ocupar

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Modelo de crecimiento lineal

$$
Y_{t} = \beta_{1} + \beta_{2}t + u_{t}
 \\
u_{t} \sim N(0,1)
$$

```python
beta1=0.15
beta2=0.45
pd_t = np.linspace(-100,100,200)
u_t = np.random.normal(0,1,200)

df_lineal = pd.DataFrame({"t":pd_t,
                          "y_t_real": beta1 + beta2 * pd_t + u_t,
                          "y_t_ajustado_real": beta1 + beta2 * pd_t})
df_lineal.head(n=15)
```

```python
plt.figure(figsize=(15,7))
plt.title("Función de crecimiento lineal",fontsize=20)

plt.scatter(df_lineal["t"], df_lineal["y_t_real"],color = "plum",label="Observaciones")
plt.plot(df_lineal["t"], df_lineal["y_t_ajustado_real"],color = "darkgreen",label="Modelo lineal ajustado real",linewidth=3)

plt.xlabel("Tiempo(años)",fontsize=15)
plt.ylabel("Crecimiento en años",fontsize=15)
plt.legend(prop={'size': 14})
plt.grid()
plt.show()
```

## Modelo de crecimiento exponencial

$$
\ln(Y_{t}) = \beta_{1} + \beta_{2}t + u_{t}
 \\
u_{t} \sim N(0,1)
$$

```python
beta1=0.25
beta2=0.06
pd_t = np.linspace(0,200,200)
u_t = np.random.normal(0,1,200)

df_exp = pd.DataFrame({"t":pd_t,
                       "y_t_real": np.exp(beta1 + beta2 * pd_t + u_t),
                       "y_t_ajustado_real": np.exp(beta1 + beta2 * pd_t)})
df_exp.head(n=15)
```

```python
plt.figure(figsize=(15,7))
plt.title("Función de crecimiento exponencial",fontsize=20)

plt.scatter(df_exp["t"], df_exp["y_t_real"],color = "lightcoral",label="Observaciones")
plt.plot(df_exp["t"], df_exp["y_t_ajustado_real"],color = "darkcyan",label="Modelo exponencial ajustado real",linewidth=3)

plt.xlabel("Tiempo(años)",fontsize=15)
plt.ylabel("Crecimiento en años",fontsize=15)
plt.legend(prop={'size': 14})
plt.grid()
plt.show()
```

## Modelo de crecimiento logístico

$$
Y_{t} = \frac{\beta_{1}}{1 + \beta_{2} e^{-\beta_{3}t}} + u_{t}
 \\
u_{t} \sim N(0,1)
$$

```python
n_data = 100
beta1 = 50
beta2 = 5
beta3 = 0.17

pd_t = np.linspace(-75,75,n_data)
u_t = np.random.normal(0,1,n_data)

df_logistico = pd.DataFrame({"t":pd_t,
                             "y_t_real": (beta1/(1 + beta2*np.exp(-1 * beta3 * pd_t))) + u_t,
                             "y_t_ajustado_real": beta1/(1 + beta2*np.exp(-1 * beta3 * pd_t))})
df_logistico.head(n=15)
```

```python
# Nota el tiempo puede ser trasladado
plt.figure(figsize=(15,7))
plt.title("Función de crecimiento logístico",fontsize=20)

plt.scatter(df_logistico["t"], df_logistico["y_t_real"],color = "goldenrod",label="Observaciones")
plt.plot(df_logistico["t"], df_logistico["y_t_ajustado_real"],color = "brown",label="Modelo logístico ajustado real",linewidth=3)

plt.xlabel("Tiempo(años)",fontsize=15)
plt.ylabel("Crecimiento en años",fontsize=15)
plt.legend(prop={'size': 14})
plt.grid()
plt.show()
```

## Modelo de crecimiento de Gompertz

$$
Y_{t} = \beta_{1}e^{-\beta_{2}e^{-\beta_{3}t}} + u_{t}
 \\
u_{t} \sim N(0,1)
$$

```python
n_data = 100
beta1 = 35
beta2 = 0.36
beta3 = 0.10

pd_t = np.linspace(-75,75,n_data)
u_t = np.random.normal(0,1,n_data)

df_gompertz = pd.DataFrame({"t":pd_t,
                            "y_t_real": (beta1*np.exp(-1*beta2*np.exp(-1*beta3*pd_t))) + u_t,
                            "y_t_ajustado_real": beta1*np.exp(-1*beta2*np.exp(-1*beta3*pd_t))})
df_gompertz.head(n=15)
```

```python
# Nota el tiempo puede ser trasladado, ¿se parece un poco al logistico no? R= Pues no
plt.figure(figsize=(15,7))
plt.title("Función de crecimiento logístico",fontsize=20)

plt.scatter(df_gompertz["t"], df_gompertz["y_t_real"],color = "lightseagreen",label="Observaciones")
plt.plot(df_gompertz["t"], df_gompertz["y_t_ajustado_real"],color = "mediumpurple",label="Modelo logístico ajustado real",linewidth=3)

plt.xlabel("Tiempo(años)",fontsize=15)
plt.ylabel("Crecimiento en años",fontsize=15)
plt.legend(prop={'size': 14})
plt.grid()
plt.show()
```
